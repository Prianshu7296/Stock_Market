```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
# File path
file_path = 'stacked_yearwise_output.xlsx'

def plot_heatmap(data):
    plt.figure(figsize=(10, 8))
    a = sns.heatmap(data,annot=True,cmap = "coolwarm",fmt='0.2f')
    plt.title("Heatmap of CP and Other Ratios")
    plt.savefig('heatmap_before_standardization.png', dpi = 300)
    plt.show()
data = pd.read_excel(file_path)
data.head()
corr_matrix = data.corr()
plot_heatmap(corr_matrix)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
#from sklearn.preprocessing import MinMaxScaler
#scaler = MinMaxScaler()
other = data.drop(columns=["CP"])
standardized_data = pd.DataFrame(scaler.fit_transform(other), columns=other.columns)
standardized_data["CP"]=data["CP"]
corr_matrix_standardized = standardized_data.corr()
plot_heatmap(corr_matrix_standardized)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pandas as pd

# Add intercept manually (important for VIF calculation)
X = standardized_data.copy()
X.dropna(inplace=True)
X['intercept'] = 1

# Automatically drop columns with VIF > 10
while True:
    # Calculate VIF for each feature
    vif_data = pd.DataFrame()
    vif_data["Feature"] = X.columns
    vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    
    # Drop the intercept
    vif_data = vif_data[vif_data['Feature'] != 'intercept']
    
    # Get the feature with the highest VIF
    max_vif = vif_data["VIF"].max()
    
    if max_vif > 10:
        # Drop the feature with the highest VIF
        feature_to_drop = vif_data.loc[vif_data["VIF"] == max_vif, "Feature"].values[0]
        X = X.drop(columns=[feature_to_drop])
        print(f"Dropped {feature_to_drop} with VIF: {max_vif}")
    else:
        break

# After the loop ends, we will have only features with VIF <= 10
X = X.drop(columns=['intercept'])

print("Remaining features after automatic VIF removal:")
print(X.columns)

# filtered_data_vif = filtered_data[features_to_keep]
# filtered_data_vif["CP"] = filtered_data["CP"]
# filtered_data_vif.dropna(inplace=True)
standardized_data = X;
corr_matrix_filtered_vif = X.corr()
plot_heatmap(corr_matrix_filtered_vif)
```

```python
def remove_outliers(df, columns, threshold=1.5):
    Q1 = df[columns].quantile(0.25)
    Q3 = df[columns].quantile(0.75)
    IQR = Q3 - Q1
    filtered_df = df[~((df[columns] < (Q1 - threshold * IQR)) | (df[columns] > (Q3 + threshold * IQR))).any(axis=1)]
    return filtered_df
filtered_data = remove_outliers(df = standardized_data, columns=standardized_data.columns, threshold=5.0)

filtered_data.dropna(inplace=True)
corr_matrix_filtered = filtered_data.corr()
plot_heatmap(corr_matrix_filtered)
```

```python
from sklearn.metrics import roc_curve, auc, precision_score, recall_score, f1_score

best_thresholds = {}  # To store best thresholds for each model

for i, (name, model) in enumerate(models.items()):
    y_probs = model.predict_proba(X_valid)[:, 1]
    thresholds = [i/100 for i in range(10, 90, 5)]  # Thresholds from 0.1 to 0.85
    best_f1 = 0
    best_thresh = 0.5  # default
    
    for thresh in thresholds:
        y_pred_thresh = (y_probs >= thresh).astype(int)  # Apply threshold
        f1 = f1_score(y_valid, y_pred_thresh)
        
        if f1 >= best_f1:
            best_f1 = f1
            best_thresh = thresh
    
    best_thresholds[name] = (best_thresh, best_f1)
    print(f"{name}: Best threshold = {best_thresh}, Best F1 Score = {best_f1:.4f}")
```

```python
for model_name in best_thresholds:
    y_probs = models[model_name].predict_proba(X_test)[:, 1]
    y_pred_thresh = (y_probs >= best_thresholds[model_name][0]).astype(int)
    print(model_name)
    print("\nClassification Report:\n", classification_report(y_test, y_pred_thresh))
    print("\nConfusion matrix\n", confusion_matrix(y_test,y_pred_thresh),"\n\n")
```
